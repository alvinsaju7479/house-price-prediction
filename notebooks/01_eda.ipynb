# üè† House Price Prediction ‚Äî EDA
This notebook explores the housing dataset and highlights patterns that influence `price`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/housing.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
plt.figure(figsize=(6,4))
df['price'].hist(bins=30)
plt.title('Target Distribution: price')
plt.xlabel('price')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
corr = df[numeric_cols].corr()['price'].sort_values(ascending=False)
corr
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['area'], df['price'], alpha=0.5)
plt.title('area vs price')
plt.xlabel('area')
plt.ylabel('price')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
df.groupby('furnishingstatus')['price'].mean().sort_values().plot(kind='bar')
plt.title('Avg price by furnishingstatus')
plt.xlabel('furnishingstatus')
plt.ylabel('avg price')
plt.tight_layout()
plt.show()
```
